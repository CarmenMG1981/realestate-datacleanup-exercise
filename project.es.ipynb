#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

Pista: Puedes ayudarte revisando la demostración multihist de Matplotlib.

```python
# TODO


# Defino una función para normalizar la variable de precios para cada población
def normalizar_precios(df):
    media = df['price'].mean()
    desviacion_estandar = df['price'].std()
    df['precio_normalizado'] = (df['price'] - media) / desviacion_estandar
    return df

# Aplico la función a cada grupo de población en 'level5'
ds_subconjunto_normalizado = ds_subconjunto.groupby('level5').apply(normalizar_precios)

# Imprimimo el dataframe resultante para verificar
print(ds_subconjunto_normalizado)


import matplotlib.pyplot as plt

# Trazo los histogramas de precios normalizados para cada población
poblaciones = ds_subconjunto_normalizado['level5'].unique()
plt.figure(figsize=(10, 6))

for poblacion in poblaciones:
    subset = ds_subconjunto_normalizado[ds_subconjunto_normalizado['level5'] == poblacion]
    plt.hist(subset['precio_normalizado'], bins=30, alpha=0.5, label=poblacion)

plt.xlabel('Precio Normalizado')
plt.ylabel('Frecuencia')
plt.title('Histogramas de Precios Normalizados por Población')
plt.legend()
plt.show()
```
